```python
import os
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.corpus import PlaintextCorpusReader
from gensim import corpora, models, similarities
import warnings
import nltk
from nltk.corpus import stopwords
import string
from nltk import sent_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer

warnings.filterwarnings("ignore")
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def read_file(file_path):
    file = open(file_path, "r") 
    file_content = file.read()
    file.close()
    return file_content
```

```python
def text_summarization(text):
    texts = [[word for word in document.lower().split() if word not in stop_words] for document in text]
    sentences = sent_tokenize(text)
    vectorizer = TfidfVectorizer()
    tfidf_vectors = vectorizer.fit_transform(sentences)
    scores = tfidf_vectors.mean(axis=1)
    top_indices = np.argsort(np.ravel(scores))[::-1]
    for i in range(5):
        print(sentences[top_indices[i]])
```

```python
data_006 = read_file('./data/bbc/tech/006.txt')
text_summarization(data_006)
```

```python
data_024 = read_file('./data/bbc/entertainment/024.txt')
text_summarization(data_024)
```
